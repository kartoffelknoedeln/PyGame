## --- Adding columns ---

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df
```

```python
# This is the same as above. The syntax [:,4:10] means you're adding all the columbs between the 4th to the 9th column.
# The sum() at the end determines if you're adding vertically or horizonally:
# 0 is vertically and 1 is horizontally.

df['Total'] = df.iloc[:,4:10].sum(axis=1)
df
```

## --- Column rearranging ---

```python
cols = list(df.columns.values)
cols_rearrange_df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
cols_rearrange_df
```

```python
index_reset = df.loc[(df['Type 1'] == 'Psychic') | (df['Type 2'] == 'Flying') & (df['Generation'] == 2)]
index_reset.reset_index(drop=True,inplace=True)
index_reset
```

## --- String search ---

## --- Regex ---

```python
# The flag 're.I' is ignore, which allows us to look for all names regardless of its cases.

import re

df.loc[df['Type 1'].str.contains('dark|steel', flags=re.I, regex=True)]
```

```python
import re

df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```
